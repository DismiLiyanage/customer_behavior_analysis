```python
import pandas as pd
from sqlalchemy import create_engine
import re

#load data
df = pd.read_csv("drive/MyDrive/customer_shopping_behavior.csv")

# This handles the exact transformation: "Purchase Amount (USD)" -> "purchase_amount"
def clean_column_names(columns):
    new_cols = []
    for col in columns:
        col = col.lower()
        # Remove parentheses and content inside (fixes "(usd)")
        col = re.sub(r'\(.*?\)', '', col)
        # Replace non-alphanumeric (spaces, dashes) with underscores
        col = re.sub(r'[^a-z0-9]+', '_', col)
        # Clean double underscores and remove trailing underscores
        col = re.sub(r'_+', '_', col).strip('_')
        new_cols.append(col)
    return new_cols

df.columns = clean_column_names(df.columns)

#feature engineering and preprocessing
# A. Fill Review Rating Nulls (Median by Category)
if 'review_rating' in df.columns:
    df['review_rating'] = df.groupby('category')['review_rating'].transform(lambda x: x.fillna(x.median()))

# B. Create Age Groups (using pd.qcut)
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)

# C. Create Purchase Frequency Days
# Fixed the spelling of "Quarterly" to match the CSV file exactly
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Quaterly': 90, # Fallback for your original code's typo
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
# Using .str.strip() ensures that hidden spaces in the CSV won't cause Null values
df['purchase_frequency_days'] = df['frequency_of_purchases'].str.strip().map(frequency_mapping)

# D. Drop Redundant Column
if 'promo_code_used' in df.columns:
    df = df.drop('promo_code_used', axis=1)

#upload to neon
conn_url = "postgresql://neondb_owner:npg_v9jC5rpRubdJ@ep-old-waterfall-ahmz04kn-pooler.c-3.us-east-1.aws.neon.tech/cu_bv?sslmode=require"

# Create the SQLAlchemy engine
engine = create_engine(conn_url)

try:
    # Use if_exists='replace' to fix the database schema with new names and features
    df.to_sql('customer', engine, if_exists='replace', index=False)
    print("--- Success! ---")
    print("1. All column names converted to lowercase snake_case.")
    print("2. Review ratings filled with category medians.")
    print("3. Age groups and frequency days columns added.")
    print("4. Data successfully uploaded to Neon.")
    print("\nFinal Column List in Database:")
    print(df.columns.tolist())
except Exception as e:
    print(f"An error occurred during upload: {e}")
```

```python
import pandas as pd
from sqlalchemy import create_engine, text
import re

# load data
df = pd.read_csv("drive/MyDrive/customer_shopping_behavior.csv")
print(f"Total rows found in CSV: {len(df)}")

#clean column names
def clean_column_names(columns):
    new_cols = []
    for col in columns:
        col = col.lower()
        col = re.sub(r'\(.*?\)', '', col)
        col = re.sub(r'[^a-z0-9]+', '_', col)
        col = re.sub(r'_+', '_', col).strip('_')
        new_cols.append(col)
    return new_cols

df.columns = clean_column_names(df.columns)

# Feature engineering
df['review_rating'] = df.groupby('category')['review_rating'].transform(lambda x: x.fillna(x.median()))
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)

freq_map = {
    'Fortnightly': 14, 'Weekly': 7, 'Monthly': 30, 'Quarterly': 90,
    'Quaterly': 90, 'Bi-Weekly': 14, 'Annually': 365, 'Every 3 Months': 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].str.strip().map(freq_map)

if 'promo_code_used' in df.columns:
    df = df.drop('promo_code_used', axis=1)

#upload with chunksize & verification
conn_url = "postgresql://neondb_owner:npg_v9jC5rpRubdJ@ep-old-waterfall-ahmz04kn-pooler.c-3.us-east-1.aws.neon.tech/cu_bv?sslmode=require"
engine = create_engine(conn_url)

try:
    print("Uploading data to Neon...")
    # method='multi' and chunksize=1000 makes it faster and more stable
    df.to_sql('customer', engine, if_exists='replace', index=False, chunksize=1000)

    # verification step-Ask the database for the count
    with engine.connect() as conn:
        result = conn.execute(text("SELECT COUNT(*) FROM customer"))
        db_count = result.scalar()
        print(f"--- SUCCESS ---")
        print(f"Rows in CSV: {len(df)}")
        print(f"Rows successfully added to Neon: {db_count}")

except Exception as e:
    print(f"--- FAILED ---")
    print(f"Error: {e}")
```
